```python
# Longer Spark operation
import time

start = time.time()
df = spark.range(10000000).repartition(100)
result = df.groupBy((df.id % 100).alias("group")).count().collect()
elapsed = time.time() - start
print(f"Processed in {elapsed:.2f} seconds")
print(f"Groups: {len(result)}")
```
