```python
import subprocess
import sys
import os

sync_dir = next((p for p in sys.path if 'jupyter_databricks_kernel' in p), None)
print(f"sync_dir: {sync_dir}")

if sync_dir:
    # uv run で httpx が import できるか確認
    result = subprocess.run(
        ["uv", "run", "python", "-c", "import httpx; print(f'httpx version: {httpx.__version__}')"],
        cwd=sync_dir,
        capture_output=True,
        text=True
    )
    print(f"uv run stdout: {result.stdout}")
    print(f"uv run stderr: {result.stderr}")
    print(f"uv run returncode: {result.returncode}")
    
    # .venv の中身確認
    venv_path = os.path.join(sync_dir, ".venv", "lib")
    if os.path.exists(venv_path):
        print(f"\n.venv/lib contents: {os.listdir(venv_path)}")
```
